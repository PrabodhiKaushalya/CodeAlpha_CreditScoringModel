```python
import nltk
from nltk.chat.util import Chat, reflections

# 1. Patterns and Responses
pairs = [
    [r"hi|hello|hey", ["Hello!", "Hey there!"]],
    [r"what is your name?", ["I am the CodeAlpha Assistant Chatbot."]],
    [r"how are you?", ["I am doing well, thank you!"]],
    [r"quit", ["Bye! Take care."]]
]

# 2. Manual Chat Loop with Input Box
def start_chat():
    chat = Chat(pairs, reflections)
    print("Chatbot started! (Type 'quit' to stop)")
    
    while True:
        # This will open a text box at the top of VS Code
        user_input = input("You: ")
        if user_input.lower() == 'quit':
            print("Chatbot: Bye! Take care.")
            break
        
        response = chat.respond(user_input)
        print(f"Chatbot: {response}")

start_chat()
```
