```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# 1. Load the pre-trained YOLOv8 model
# This will automatically download the model file on the first run
model = YOLO('yolov8n.pt') 

# 2. Run object detection on your image
results = model('test_image1.jpg')

# 3. Show the result
for r in results:
    im_array = r.plot()  # plot a BGR numpy array of predictions
    im_rgb = cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB) # Convert to RGB for plotting
    
    plt.figure(figsize=(10,10))
    plt.imshow(im_rgb)
    plt.axis('off')
    plt.show()

print("Object detection complete!")
```
